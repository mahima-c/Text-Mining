```python
# write a function to expand contraction in the text
def expand_contractions(sent,mapping):
#     pattern for matching contraction with their expansion
    pattern=re.compile('({})'.format('|'.join(mapping.keys())),flags=re.IGNORECASE|re.DOTALL)
    def expand_map(contraction):
        
#         using group method to asscess of subgroup of match
        match=contraction.group(0)
        print(match)
#         print("1")
#         to retain the correct case of world
        first_char = match[0]
#         find out the expantion
        expansion=mapping.get(match) if mapping.get(match) else mapping.get(match.lower())
    
        expansion=first_char + expansion[1:]
        return expansion
#     using sub method to replace all contraction with their expantion
# for expand_amp will be called for every non-overlapping occurance of pattern
    expand_sent=pattern.sub(expand_map, sent)#expand_map is apply on sent
    return expand_sent
```

```python
# corrected word #
# spelling mistake and repetative word
# repetative character

# remove repetative char from token
from nltk.corpus import wordnet
def remove_repeated_char(tokens):
#     char that occure twice among their char
    pattern=re.compile(r'(\w*)(\w)\2(\w*)')
    rep1=r'\1\2\3'
    def replace(tkn):
#         check for sematically correct word
        if wordnet.synsets(tkn):
            return tkn
#         replace the token
        tkn_c=pattern.sub(rep1,tkn)
#     recursive call
        return replace(tkn_c) if tkn_c != tkn else tkn_c

#     correct each token
    token_c=[replace(tn) for tn in tokens]
    return token_c

```

```python
# write a function  to combine or chain  all thr tranformation step for a text corpus
# includes expanding contraction ,postagging, lemmatization
# special character removal,stopword removal,tokenmization

import re
import string
import nltk
# load all the english stopwords
seq_sw=nltk.corpus.stopwords.words('english')

# load the contraction mapping
# contraction map

# load lemmatization
from nltk.stem import WordNetLemmatizer
wnl1=WordNetLemmatizer()

# define function to tokenize text into tokens
def tokenize_text(text):
    tokens=nltk.word_tokenize(text)
    tokens=[tn.strip() for tn in tokens]
    return tokens

# define a function for expanding contraction
def expand_contractions(sent,mapping):
    
    pattern=re.compile('({})'.format('|'.join(mapping.keys())),flags=re.IGNORECASE|re.DOTALL)
    def expand_map(contraction):
        
        match=contraction.group(0)

        first_char = match[0]
        expansion=mapping.get(match) if mapping.get(match) else mapping.get(match.lower())
    
        expansion=first_char + expansion[1:]
        return expansion
    
    expand_sent=pattern.sub(expand_map, sent)#expand_map is apply on sent
    return expand_sent

# define a function for lemmatization of text based on pos_tag
from nltk.corpus import wordnet as wn

# annonate text token with pos tags
def pos_tag_text(text):
    
    #   convert penn treebank tag to wordnet tag
    # since the wordnetlemmatization checks for pos tag annotation based on wordnet from
    def penn_to_wn_tags(pos_tag):
        if pos_tag.startswith('J'):
            return wn.ADJ
        elif pos_tag.startswith('V'):
            return wn.VERB
        elif pos_tag.startswith('N'):
            return wn.NOUN
        elif pos_tag.startswith('R'):
            return wn.ADV
        else:
            return None
        
    # tokenize text into token using tokenixe_text function
    tokens=tokenize_text(text)
    
    # using pos_tag function to annonatee post tags
    tokens_t=nltk.pos_tag(tokens,tagset='universal')
    
    #convert the pos tag from penn treebank syntax to wordnet
    tokens_p=[(word.lower(),penn_to_wn_tags(pos_tag)) for word,pos_tag in tokens_t]
    
    return tokens_p


# lemmatize text based on pos tags
def lemmatize_text(text):
    #pos tagging
    token_p=pos_tag_text(text)
    
    #lemmatize 
    token_lm=[wnl1.lemmatize(word) if pos_tag else word for word,pos_tag in token_p]
    #concatenate tokens
    text_lm=' '.join(token_lm)
    return text_lm


# define a function remove special char and symbol
def remove_special_char(text):
    #tokenize text into token using tokenize_text function
    tokens=tokenize_text(text)
    pattern==re.compile('[{}]'.format(re.escape(string.punctuation)))
    tokens_f=filter(None,[pattern4.sub('',tn) for tn in tokens])
    text_f=' '.join(tokens_f)
    return text_f

# define a function to remove stopword form text data
def remove_stopwords(text):
    tokens=tokenize_text(text)
    tokens_f=[tn for tn in tokens if tn not in seq_sw]
    text_f=' '.join(tokens_f)
    return text_f

# define a function to chain all the transformation steps
def transform_corpus(corpus,tokenize=False):
    corpus_t=[]
    for text in corpus:
        text=expand_contractions(text,contraction_map)
        text=lemmatize_text(text)
        text=remove_repeated_char(text)
        text=remove_stopwords(text)
        corpus_t.append(text)
        
        if tokenize:
            text=tokenize_text(text)
            corpus_t.append(text)
            
    return corpus_t
```

```python
# #bag of word model
# simplest vectorization model
# define a function to create a bag of word feature extraction or vectorization model
from sklearn.feature_extraction.text import CountVectorizer
def  bow_extractor(corpus,ngram_range=(1,1)):
#     using a instance of countvectorizer class to define the type of feature to be extracted
# first arg:term having a minimum freq of (default)
# second arg:default unigrams only :1 mean uni,2 means bi,and
# range of(1,3) would mean uni to tri (uni,bi,tri)
    vectorizer=CountVectorizer(min_df=1,ngram_range=ngram_range)
#     using fit_transform method to extract the features from corpus(term-document matrix)
    features=vectorizer.fit_transform(corpus)
    return vectorizer,features
```
